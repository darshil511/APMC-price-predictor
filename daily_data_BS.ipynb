```python
from bs4 import BeautifulSoup
import requests
```

```python
html_text = requests.get('https://apmcgondal.scmsolution.in/Daily_Rate.aspx').text
# print(html_text)
soup = BeautifulSoup(html_text, 'html.parser')
table = soup.find('table', class_ = 'table table-condensed')
```

```python
data = []
headings = table.find_all("th")
if headings:
    heading_data = [heading.get_text(strip=True) for heading in headings]
data.append(heading_data)

# Iterate over each row in the table
for row in table.find_all("tr"):
    # Find all cells (td) in the row
    cells = row.find_all("td")
    
    # Ensure that there are cells in the row
    if cells:
        # Extract text from each cell and append to the data list
        row_data = [cell.get_text(strip=True) for cell in cells]
        data.append(row_data)

# Print the extracted data for debugging
for row in data:
    print(row)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(data)
df.columns = df.iloc[0]
df
```

```python
df = df.drop(0)
df
```
